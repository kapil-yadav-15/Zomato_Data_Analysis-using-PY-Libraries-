```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading of Dataset

```python
dataframe = pd.read_csv("C:\\Users\\deepa\\Downloads\\Zomatodata.csv")
```

# Q1: Type of restaurant most ordered from


```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

# Majority of the restaurantfalls in dining category

# Q2: Votes received by each restaurant type

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})

plt.plot(result, c='green', marker='o')
plt.xlabel("Type of restaurant", c='red', size=20)
plt.ylabel("Votes", c='red', size=20)
plt.show
```

# Dining restaurants have recieved maximum votes 

# Q3: Ratings received by majority of restaurants


  ### Convert the data type of column - rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'], bins=5)
plt.title("ratings distribution")
plt.show()
```

## Majority restaurants got ratings between 3.5 to 4.5

# Q4: Average spending of couples per order

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

### Majority of couples prefer restaurants with an approx cost of 300 rupees

# Q5: Mode receiving maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

### Offline orders recieved less rating compared to online orders
